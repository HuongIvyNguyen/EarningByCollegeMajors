```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline 

recent_grads = pd.read_csv("recent-grads.csv")
first_row = recent_grads.iloc[0]
print(first_row)
```

```python
#Generating scatter plot between sample_size and median
ax1=recent_grads.plot(x="Sample_size",y="Median",kind='scatter')
ax1.set(title='Sample size versus Median',xlabel='Sample_size',ylabel='Median')
plt.show()
```

```python
#Generating scatter plot between sample_size and Unemployment_rate
ax2=recent_grads.plot(x="Sample_size",y="Unemployment_rate",kind='scatter')
ax2.set(title='Sample size versus Unemployment Rate',xlabel='Sample_size',ylabel='Unemployment rate')
plt.show()
```

```python
#Generating scatter plot between Full time and Median
ax3=recent_grads.plot(x="Full_time",y="Median",kind='scatter')
ax3.set(title='Full time versus Median',xlabel='Full_time',ylabel='Median')
plt.show()
```

```python
#Generating scatter plot between ShareWomen and Unemployment_rate
ax4=recent_grads.plot(x="ShareWomen",y="Unemployment_rate",kind='scatter')
ax4.set(title='ShareWomen versus Unemployment Rate',xlabel='ShareWomen',ylabel='Unemployment rate')
plt.show()
```

```python
#Generating scatter plot between Men and Median
ax5=recent_grads.plot(x="Men",y="Median",kind='scatter')
ax5.set(title='Men versus Median',xlabel='Men',ylabel='Median')
plt.show()
```

```python
#Generating scatter plot between Women and Median
ax6=recent_grads.plot(x="Women",y="Median",kind='scatter')
ax6.set(title='Women versus Median',xlabel='Women',ylabel='Median')
plt.show()
```
